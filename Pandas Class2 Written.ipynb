```python
df_sales = pd.read_excel('E:\Datascience\Pandas\Datasets\Superstore_USA.xlsx')
```

```python
df_sales.head()
```

```python
def profit_flag(a):
    if a >= 0:
        return 'positive'
    else:
        return 'negative'
```

```python
profit_flag(-45)
```

```python
profit_flag(50)
```

```python
df_sales['flag_profit'] = df_sales['Profit'].apply(profit_flag)
```

```python
df_sales.head()
```

```python
df_sales['len_cust_name'] = df_sales['Customer Name'].apply(len)
```

```python
df_sales
```

```python
def quantity_flag(a):
    if a < 10:
        return 'low'
    elif a > 10 and a < 20:
        return 'medium'
    else :
        return 'high'
```

```python
df_sales['flag_quantity'] = df_sales['Quantity ordered new'].apply(quantity_flag)
```

```python
df_sales.tail()
```

```python
df_sales['square_quantity'] = df_sales['Quantity ordered new'].apply(lambda a: a**2)
```
